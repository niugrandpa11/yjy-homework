**最小生成树（Minimum Spanning Tree）**

设  **G**  为一个带权重的连通无向图，且所有边的权重均不相等。令$e_i$ 为权重第 $i$ 小的边。最小生成树（MST）是否必须包含 $e_1$ ? 同理，是否必须包含 $e_2$ 和 $e_3$ ? 若必须包含，请给出证明；否则，请构造反例。需从基本原理论证，不能依赖割引理(cut lemma) 或 Prim/Kruskal算法的正确性。


最小生成树是图论中的一个关键概念，指的是在无向图中，连接所有顶点且不形成任何环的树，同时确保所有边的权重之和达到最小值。构建最小生成树的常用算法包括：
Prim算法：从一个选定的顶点开始，逐步向树中添加边，每次添加连接树与非树顶点且权重最小的边。
Kruskal算法：先将图中的所有边按权重从小到大排序，然后依次选择边加入生成树，只要这条边不会与已选择的边形成环。

```python
class UnionFind:
    def __init__(self, size):
        self.parents = list(range(size))
        self.ranks = [0] * size

    def root(self, i):
        if self.parents[i] != i:
            self.parents[i] = self.root(self.parents[i])
        return self.parents[i]

    def unite(self, i, j):
        ri, rj = self.root(i), self.root(j)
        if ri == rj:
            return False
        if self.ranks[ri] < self.ranks[rj]:
            ri, rj = rj, ri
        self.parents[rj] = ri
        if self.ranks[ri] == self.ranks[rj]:
            self.ranks[ri] += 1
        return True

def kruskal(edges, num_vertices):
    edges.sort(key=lambda edge: edge[2])
    uf = UnionFind(num_vertices)
    selected_edges = []
    for u, v, weight in edges:
        if uf.unite(u, v):
            selected_edges.append((u, v, weight))
    return selected_edges

def edge_in_mst(edges, num_vertices, target_edge):
    mst_edges = kruskal(edges, num_vertices)
    mst_edge_set = {(min(u, v), max(u, v)) for u, v, _ in mst_edges}

    u, v, _ = target_edge
    target = (min(u, v), max(u, v))
    in_mst = target in mst_edge_set

    remaining_edges = [e for e in edges if (min(e[0], e[1]), max(e[0], e[1])) != target]
    uf = UnionFind(num_vertices)
    min_crossing_weight = float('inf')
    for e in remaining_edges:
        if uf.root(e[0]) != uf.root(e[1]):
            min_crossing_weight = min(min_crossing_weight, e[2])
        uf.unite(e[0], e[1])

    definitely_in = _ <= min_crossing_weight if min_crossing_weight != float('inf') else True

    if not in_mst:
        definitely_not_in = False  # Requires more complex cycle detection logic
    else:
        definitely_not_in = False  # If already in MST, cannot be certain it's not

    if definitely_in:
        return "YES"
    if definitely_not_in:
        return "NO"
    return "MAYBE"

def solve(n, m, edges, queries):
    results = []
    for query_edge in queries:
        result = edge_in_mst(edges, n, query_edge)
        results.append(result)
    return results

# Test the function
n, m = 4, 5
edges = [(0, 1, 1), (1, 2, 2), (2, 3, 3), (3, 0, 4), (0, 2, 5)]
queries = [(0, 1, 1), (1, 2, 2), (0, 2, 5)]
results = solve(n, m, edges, queries)
for i, result in enumerate(results, 1):
    print(f"Edge e{i}: {result}")
```

## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


```python
import heapq
from collections import defaultdict

def can_reach_target(start, end, max_path_length, graph):
    """
    判断从起点到终点是否存在一条路径，其长度不超过指定的最大长度
    
    参数:
    start - 起点
    end - 终点
    max_path_length - 允许的最大路径长度
    graph - 图的邻接表示，格式为 {vertex: [(neighbor, weight), ...]}
    
    返回:
    布尔值，表示是否存在满足条件的路径
    """
    # 初始化优先队列，存储(当前路径长度, 当前顶点)
    priority_queue = [(0, start)]
    
    # 记录到各顶点的最短路径长度
    shortest_paths = {vertex: float('inf') for vertex in graph}
    shortest_paths[start] = 0
    
    while priority_queue:
        current_length, current_vertex = heapq.heappop(priority_queue)
        
        # 如果当前路径已经超过了最大允许长度，跳过
        if current_length > max_path_length:
            continue
            
        # 如果找到了到终点的有效路径，返回True
        if current_vertex == end:
            return True
            
        # 如果当前路径不是到该顶点的最短路径，跳过
        if current_length > shortest_paths[current_vertex]:
            continue
            
        # 遍历所有邻居顶点
        for neighbor, edge_weight in graph[current_vertex]:
            new_length = current_length + edge_weight
            
            # 如果找到更短的路径，更新并加入队列
            if new_length < shortest_paths[neighbor]:
                shortest_paths[neighbor] = new_length
                heapq.heappush(priority_queue, (new_length, neighbor))
    
    # 如果队列排空仍未找到路径，则返回False
    return False

# 示例用法
def main():
    # 构建图的邻接列表
    num_vertices = 4
    graph = defaultdict(list)
    
    # 确保所有顶点都被初始化（即使没有出边）
    for vertex in range(num_vertices):
        graph[vertex]
    
    # 添加边
    edges = [
        (0, 1, 1),   # 0->1，权重1
        (1, 2, 2),   # 1->2，权重2
        (2, 3, 3),   # 2->3，权重3
        (0, 2, 5),   # 0->2，权重5
        (1, 3, 6)    # 1->3，权重6
    ]
    
    for src, dest, weight in edges:
        graph[src].append((dest, weight))
    
    # 参数设置
    start_vertex = 0
    end_vertex = 3
    max_allowed_length = 4
    
    # 执行判断
    is_reachable = can_reach_target(start_vertex, end_vertex, max_allowed_length, graph)
    
    print(f"从顶点{start_vertex}到顶点{end_vertex}是否有路径长度不超过{max_allowed_length}:", is_reachable)

if __name__ == "__main__":
    main()
```

## 问题 4

**逃离问题**

一个 $ n \times n $ 的网格是一个无向图，由 $ n $ 行和 $ n $ 列的顶点组成，如下图所示。我们用 $(i,j)$ 表示第 $ i $ 行和第 $ j $ 列的顶点。除了边界顶点，网格中的所有顶点都有四个邻居，即满足 $ i = 1, i = n, j = 1 $ 或 $ j = n $ 的点 $(i,j)$。

给定网格中的 $ m \leq n^2 $ 个起点 $(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$，逃离问题是确定是否存在 $ m $ 条顶点不相交的路径（即路径之间不相交），从这些起点到边界上的任意 $ m $ 个不同点。例如，图1中的网格存在逃离。

(1) 该问题可以看作是一个最大流问题。考虑一个流网络，其中顶点和边都有容量。也就是说，进入任何给定顶点的总正流量受到容量限制。证明在具有边和顶点容量的网络中确定最大流可以简化为在具有可比大小的普通流网络上的最大流问题。更准确地说，你需要将一个具有顶点和边容量的网络 $ G = (V,E) $ 转换为另一个仅具有边容量的网络 $ G' = (V', E') $，使得两个网络上的最大流相同，并且你构建的新网络具有 $ V' = O(V) $ 个顶点和 $ E' = O(E) $ 条边。你可以假设网络是连通的。

(2) 描述一个解决逃离问题的高效算法，并分析其运行时间。

```python
from collections import deque
import heapq

def idx(i, j, n=6):
    return i * n + j

def bfs(graph, src, tgt, parents):
    visited = [False] * len(graph)
    queue = deque([src])
    visited[src] = True
    while queue:
        u = queue.popleft()
        for v, cap in enumerate(graph[u]):
            if not visited[v] and cap > 0:
                queue.append(v)
                visited[v] = True
                parents[v] = u
    return visited[tgt]

def ff(graph, src, tgt):
    parents = [-1] * len(graph)
    max_flow = 0
    paths = []

    while bfs(graph, src, tgt, parents):
        path_flow = float('inf')
        s = tgt
        path = []
        while s != src:
            path.append(s)
            path_flow = min(path_flow, graph[parents[s]][s])
            s = parents[s]
        path.append(src)
        path.reverse()
        paths.append(path)

        v = tgt
        while v != src:
            u = parents[v]
            graph[u][v] -= path_flow
            graph[v][u] += path_flow
            v = u
        max_flow += path_flow

    return max_flow, paths

n = 6
V = n * n + 2
source, sink = V - 2, V - 1
graph = [[0] * V for _ in range(V)]

starts = [(1,1), (3,1), (5,1), (0,2), (1,2), (3,2), (5,2), (1,3), (3,3), (5,3)]
boundary_starts = [(0,1), (5,1), (5,2), (5,3)]

for i, j in starts:
    graph[source][idx(i, j)] = 1

for i in range(n):
    for j in range(n):
        idx_val = idx(i, j)
        for di, dj in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            ni, nj = i + di, j + dj
            if 0 <= ni < n and 0 <= nj < n:
                graph[idx_val][idx(ni, nj)] = 1

boundary_points = set()
for i in range(n):
    boundary_points.add(idx(0, i))
    boundary_points.add(idx(n-1, i))
    boundary_points.add(idx(i, 0))
    boundary_points.add(idx(i, n-1))
occupied_boundary = {idx(i, j) for i, j in boundary_starts}
for bp in boundary_points:
    if bp not in occupied_boundary:
        graph[bp][sink] = 1

max_flow, paths = ff(graph, source, sink)

print(f"需要逃离的点数: {len(starts)}")
print(f"需要逃离的点: {[(i, j) for i, j in starts]}")
print(f"最大流: {max_flow}")
print(f"是否所有需要逃离的点都能到达边界: {max_flow == len(starts)}")
print("\n逃离路径（顶点编号及对应坐标）：")
for i, path in enumerate(paths, 1):
    path_coords = [(v // n, v % n) if v < n * n else v for v in path]
    print(f"路径 {i}: {path} (坐标: {path_coords})")
print("\n已经在边界上的点（无需逃离）：")
for i, j in boundary_starts:
    idx_val = idx(i, j)
    print(f"点 ({i}, {j})，编号: {idx_val}")
```
